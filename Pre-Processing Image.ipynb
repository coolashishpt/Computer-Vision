```python
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image

# Just for checking directory
source_path = "data/voc"
classes = os.listdir(source_path)
classes
```

```python
def images_in_list(src_path):
    import matplotlib.pyplot as plt
    import os
    images = []
    image_label = []
#     image_with_target = dict()
#     source_path = "data/voc"
    classes = os.listdir(source_path)
    for class_name in classes:
        path = source_path+"/"+class_name
        images_path = os.listdir(path)
        for images_name in images_path:
            images_path = path+"/"+images_name
            image = plt.imread(images_path)
            images.append(image)
            image_label.append(class_name)
#             image_with_target[class_name] = image
    return images, image_label
```

```python
image, image_label = images_in_list("data/voc")
```

```python
indx = 160
print("Image Label:",image_label[indx])
plt.imshow(image[indx])
plt.show()
```

```python
def resized_image(image_list, size=(128,128), bg_color="white"):
    resized_images = list()
    for image in image_list:
        src_image = Image.fromarray(image)
        src_image.thumbnail(size, Image.ANTIALIAS)
        # Create a new square background image
        new_image = Image.new("RGB", size, bg_color)
        # Paste the resized image into the center of the square background
        new_image.paste(src_image, (int((size[0] - src_image.size[0]) / 2), int((size[1] - src_image.size[1]) / 2)))
        
        resized_images.append(new_image)
    return resized_images
```

```python
new_image = resized_image(image)
```

```python
fig = plt.figure(figsize=(15,15))

indx = 100
a = fig.add_subplot(3,3,1)
imgplot = plt.imshow(image[indx])
a.set_title("Original Image")

a = fig.add_subplot(3,3,2)
imgplot = plt.imshow(new_image[indx])
a.set_title("Resized New Image")

plt.show()
```
